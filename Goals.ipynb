```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from CSV
data = pd.read_csv('C:/Users/S558637/Downloads/BigDataProjectDataset/data.csv', encoding='latin1')

# Convert 'InvoiceDate' to datetime
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])

# Purchase frequency
purchase_frequency = data.groupby('CustomerID')['InvoiceNo'].nunique()

# Popular products
popular_products = data['Description'].value_counts().head(10)

# Seasonal trends
data['Month'] = data['InvoiceDate'].dt.month
monthly_sales = data.groupby('Month')['Quantity'].sum()

# Visualization of purchase frequency
plt.figure(figsize=(10,6))
purchase_frequency.plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
plt.title('Purchase Frequency Distribution')
plt.xlabel('Number of Purchases')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Visualization of popular products
plt.figure(figsize=(10,6))
popular_products.plot(kind='bar', color='skyblue')
plt.title('Top 10 Popular Products')
plt.xlabel('Product')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualization of seasonal trends
plt.figure(figsize=(10,6))
monthly_sales.plot(kind='line', marker='o', color='skyblue')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Quantity Sold')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the dataset into a pandas DataFrame
data = pd.read_csv(r'C:\Users\S558637\Downloads\archive\commerce_data.csv')

# Display basic information about the DataFrame
print(data.info())

# Display the first few rows of the DataFrame
print(data.head())

# Convert the 'time' column to datetime type
data['time'] = pd.to_datetime(data['time'])

# Extract year, month, and day from the 'time' column
data['year'] = data['time'].dt.year
data['month'] = data['time'].dt.month
data['day'] = data['time'].dt.day

# Identify purchase frequency
purchase_frequency = data[data['event_name'] == 'purchase'].groupby(['year', 'month']).size()

# Plot purchase frequency
purchase_frequency.plot(kind='bar', figsize=(10, 6))
plt.title('Purchase Frequency Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Identify popular products
popular_products = data[data['event_name'] == 'purchase']['product_id'].value_counts().head(10)

# Plot popular products
popular_products.plot(kind='bar', figsize=(10, 6))
plt.title('Top 10 Popular Products')
plt.xlabel('Product ID')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Identify seasonal trends
seasonal_trends = data.groupby(['month']).size()

# Plot seasonal trends
seasonal_trends.plot(kind='line', figsize=(10, 6))
plt.title('Seasonal Trends')
plt.xlabel('Month')
plt.ylabel('Number of Events')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid()
plt.show()

# Select relevant features for segmentation
selected_features = ['user_id', 'product_id', 'price']

# Group data by user_id and aggregate on selected features
customer_data = data.groupby('user_id')[selected_features].agg({
    'product_id': 'nunique',  # Number of unique products purchased
    'price': 'sum'  # Total amount spent by the customer
})

# Normalize the data
scaler = StandardScaler()
normalized_data = scaler.fit_transform(customer_data)

# Determine the optimal number of clusters using the Elbow method
wcss = []  # Within-cluster sum of squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(normalized_data)
    wcss.append(kmeans.inertia_)

# Plot the Elbow method to find the optimal number of clusters
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

# Based on the Elbow method, select the optimal number of clusters and perform clustering
optimal_num_clusters = 3  # Adjust this based on the Elbow plot
kmeans = KMeans(n_clusters=optimal_num_clusters, init='k-means++', random_state=42)
customer_data['cluster'] = kmeans.fit_predict(normalized_data)

# Analyze the clusters
cluster_means = customer_data.groupby('cluster').mean()
cluster_counts = customer_data['cluster'].value_counts()

# Print cluster information
print("Cluster Information:")
for cluster in range(optimal_num_clusters):
    print(f"Cluster {cluster}:")
    print(f"Number of customers: {cluster_counts[cluster]}")
    print(f"Average number of unique products purchased: {cluster_means.loc[cluster, 'product_id']:.2f}")
    print(f"Average total amount spent: ${cluster_means.loc[cluster, 'price']:.2f}\n")
```
